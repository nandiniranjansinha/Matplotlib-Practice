```python
plt.figure(figsize = (8,4))
plt.title('Gas Prices over time (in USD)')

for country in gas:
  if country != 'Year':
    plt.plot(gas['Year'], gas[country], marker = '^')

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.title('Gas Prices over time (in USD)')

for country in gas:
  if country != 'Year':
    plt.plot(gas['Year'], gas[country], marker = '^', label = country)

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend()
plt.show()
```

```python
# For a few Countries

plt.figure(figsize = (10,8))
plt.title('Gas Prices over time (in USD)')

countries = ['UK', 'Australia', 'Italy', 'Japan']
for country in countries:
  if country != 'Year':
    plt.plot(gas['Year'], gas[country], marker = '^', label = country)

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend()
plt.show()
```

```python
# Adding year 2011 in x-axis

# For a few Countries

plt.figure(figsize = (10,8))
plt.title('Gas Prices over time (in USD)')

countries = ['UK', 'Australia', 'Italy', 'Japan']
for country in countries:
  if country != 'Year':
    plt.plot(gas['Year'], gas[country], marker = '^', label = country)

plt.xticks(gas.Year[::3].tolist() + [2011])
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend()

plt.savefig('Gas_Price_fig.png', dpi = 300)

plt.show()
```

## Pie Chart

```python
fifa['Preferred Foot']
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

print("Players who use left foot the most: ", left)
print("Players who use right foot the most: ", right)
```

```python
plt.pie([left,right])
plt.show()
```

```python
labels = ['Left','Right']
colors = ['#F49BAB','#9B7EBD']


plt.pie([left,right], labels = labels, colors = colors)

plt.legend()
plt.show()
```

```python
# Percentage of each

labels = ['Left','Right']
colors = ['#F49BAB','#9B7EBD']


plt.pie([left,right], labels = labels, colors = colors, autopct='%.2f') # till 2 digits after decimal

plt.legend()
plt.show()
```

```python
# Adding % sign

labels = ['Left','Right']
colors = ['#F49BAB','#9B7EBD']


plt.pie([left,right], labels = labels, colors = colors, autopct='%.2f %%') # till 2 digits after decimal

plt.legend()
plt.show()
```

```python
# Adding % sign

labels = ['Left','Right']
colors = ['#F49BAB','#9B7EBD']

plt.title("Foot Preference of FIFA Players")

plt.pie([left,right], labels = labels, colors = colors, autopct='%.2f %%')

plt.legend()
plt.show()
```

```python
# FIFA Weight Pie Chart

fifa.Weight
```

```python
# To make a pie chart, the weight should have numbers only not string
# We need to remove lbs

fifa.Weight = [x.strip('lbs') if type(x) == str else x for x in fifa.Weight]

fifa.Weight
```

```python
fifa.Weight[0]
```

```python
# Still a string

fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

fifa.Weight
```

```python
fifa.Weight[0]
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

print("Light: ", light)
print("Light Medium: ", light_medium)
print("Medium: ", medium)
print("Medium Heavy: ", medium_heavy)
print("Heavy: ", heavy)
```

```python
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-149','150-174','175-199','200 and Above']
colors = ['#90D1CA','#E55050','#FBDB93','#FFDBDB','#5F8B4C']
plt.figure(figsize=(10,10))
plt.pie(weights, labels = labels, colors = colors, autopct = '%.2f %%')
plt.legend()
plt.show()
```

```python
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-149','150-174','175-199','200 and Above']

plt.style.use('default')

plt.figure(figsize=(10,10))
plt.pie(weights, labels = labels, autopct = '%.2f %%')
plt.legend()
plt.show()
```

```python
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-149','150-174','175-199','200 and Above']

plt.style.use('ggplot')

plt.figure(figsize=(10,10))
plt.pie(weights, labels = labels, autopct = '%.2f %%')
plt.legend()
plt.show()
```

```python
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-149','150-174','175-199','200 and Above']
colors = ['#90D1CA','#E55050','#FBDB93','#FFDBDB','#5F8B4C']
plt.figure(figsize=(10,10))
plt.pie(weights, labels = labels, colors = colors, autopct = '%.2f %%', pctdistance = 0.8)
plt.legend()
plt.show()
```

```python
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-149','150-174','175-199','200 and Above']
colors = ['#90D1CA','#E55050','#FBDB93','#FFDBDB','#5F8B4C']
plt.figure(figsize=(10,10))

explode = (.1,.1,.1,.1,.1)

plt.pie(weights, labels = labels, colors = colors, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)
plt.legend()
plt.show()
```

```python
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-149','150-174','175-199','200 and Above']
colors = ['#90D1CA','#E55050','#FBDB93','#FFDBDB','#5F8B4C']
plt.figure(figsize=(10,10))

explode = (.5,0,0,0,.5)

plt.pie(weights, labels = labels, colors = colors, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,10), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']
labels=['FC Barcelona','Real Madrid','NE Revolution']
boxes = plt.boxplot([barcelona, madrid, revs])

for box in boxes['boxes']:
  box.set(color = '#4286f4', linewidth = 2)


plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```

```python
plt.figure(figsize=(6,10), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']
labels=['FC Barcelona','Real Madrid','NE Revolution']
boxes = plt.boxplot([barcelona, madrid, revs], labels = labels, patch_artist = True )

for box in boxes['boxes']:
  box.set(color = '#4286f4', linewidth = 2)
  box.set(facecolor = '#e0e0e0') # need to set patch_artist = True

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

# medianprops -> same thickness

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')

plt.show()
```

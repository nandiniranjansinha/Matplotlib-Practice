# What is DPI in matplotlib?
DPI defines how many pixels per inch your figure has.

Higher DPI = higher resolution (clearer image, bigger file size).

Lower DPI = lower resolution (blurry or pixelated image).

````python
x1 = [1,2,3,4,5]
y = [2,4,6,8,10]
x2 = np.arange(0,5,0.5)

plt.figure(figsize = (5,3), dpi = 300)

plt.title("My First Graph", fontdict={'fontname': comicNeude_Reg, 'fontsize' : 16 })
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xticks([1,2,3,4,5])
#plt.yticks([2,4,6,8,10])
plt.plot(x1,y, label = '2x', color = 'purple', linewidth = 4, linestyle = "--" ,marker = '*', markersize = 14, markeredgecolor = 'pink')
plt.plot(x2[:5], x2[:5]**2, label = "Sqaured Function", color = 'cyan', linewidth = 2, linestyle = '--', marker = "^", markersize = 10, markeredgecolor = 'orange')
plt.plot(x2[4:], x2[4:]**2, color = 'cyan', linewidth = 2, linestyle = '-', marker = "^", markersize = 10, markeredgecolor = 'orange')
plt.legend()
plt.show()
````

````python
# To save the graph
x1 = [1,2,3,4,5]
y = [2,4,6,8,10]
x2 = np.arange(0,5,0.5)

plt.figure(figsize = (5,3), dpi = 300)

plt.title("My First Graph", fontdict={'fontname': comicNeude_Reg, 'fontsize' : 16 })
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xticks([1,2,3,4,5])
#plt.yticks([2,4,6,8,10])
plt.plot(x1,y, label = '2x', color = 'purple', linewidth = 4, linestyle = "--" ,marker = '*', markersize = 14, markeredgecolor = 'pink')
plt.plot(x2[:5], x2[:5]**2, label = "Sqaured Function", color = 'cyan', linewidth = 2, linestyle = '--', marker = "^", markersize = 10, markeredgecolor = 'orange')
plt.plot(x2[4:], x2[4:]**2, color = 'cyan', linewidth = 2, linestyle = '-', marker = "^", markersize = 10, markeredgecolor = 'orange')
plt.legend()

plt.savefig('my_first_graph.png', dpi = 300)

plt.show()
````

If Image('my_first_graph.png') shows a blank white canvas, it usually means the image file was saved before the plot was actually drawn or the plot was cleared before saving.

**How to fix it:**

Make sure you save the figure after plotting and before plt.show(), or dont call plt.show() before saving.

**Why?**

plt.show() may clear the figure, so if you save after plt.show(), the file might be empty or blank.

***Always save before plt.show().***

### Bar Chart

````python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.bar(labels, values)
plt.show()
````

````python
# To resize

labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize = (6,6))

plt.bar(labels, values)
plt.show()
````

````python
# To add patterns/designs to each bar

labels = ['A', 'B', 'C']
values = [1,4,2]

bar = plt.bar(labels, values)
bar[0].set_hatch('/')
bar[1].set_hatch('*')
bar[2].set_hatch('+')

plt.figure(figsize = (6,6))
plt.show()
````

````python
labels = ['A', 'B', 'C']
values = [1,4,2]

bar = plt.bar(labels, values)
bar[0].set_hatch('|')
bar[1].set_hatch('=')
bar[2].set_hatch('-')

plt.figure(figsize = (6,6))
plt.show()
````

````python
labels = ['A', 'B', 'C']
values = [1,4,2]

bar = plt.bar(labels, values)
bar[0].set_hatch('*+-')
bar[1].set_hatch('*+')
bar[2].set_hatch('*-')

plt.figure(figsize = (6,6))
plt.show()
````

````python
# To add patterns using for loop

labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)
patterns = ['/','O','*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))

plt.figure(figsize = (6,6))
plt.show()
````
